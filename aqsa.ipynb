##### reading pptx files

```python
from pptx import Presentation
pptx_file ='MARGIE’S.pptx'
# Load the presentation
presentation = Presentation(pptx_file)

# Loop through slides
for slide_num, slide in enumerate(presentation.slides):
    print(f"Slide {slide_num + 1}:")
    
    # Loop through shapes in the slide
    for shape in slide.shapes:
        # Check if the shape contains text
        if hasattr(shape, "text"):
            print(f"  Text: {shape.text}")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
```

##### extract diff types of shapes

```python
for slide in presentation.slides:
    for shape in slide.shapes:
        if shape.has_table:
            table = shape.table
            for row in table.rows:
                for cell in row.cells:
                    print(cell.text)
```

##### handling images

```python
from pptx.enum.shapes import MSO_SHAPE_TYPE

for slide in presentation.slides:
    for shape in slide.shapes:
        if shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
            image = shape.image
            image_bytes = image.blob  # Get the image bytes
            with open("image.jpg", "wb") as f:
                f.write(image_bytes)
```

##### Example code to parse and extract data

```python
from bs4 import BeautifulSoup

# Open the HTML file and parse it using BeautifulSoup
with open("data.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Example: Extract all the <h1> headers
for h1_tag in soup.find_all("h1"):
    print(h1_tag.text)

# Example: Extract all links (<a> tags)
for link in soup.find_all("a"):
    print(link.get("href"))
```

#### Example code to parse HTML with lxml

```python
from lxml import html

# Open and read the HTML file
with open("data.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content
tree = html.fromstring(html_content)

# Example: Extract all <h1> tags
h1_tags = tree.xpath("//h1/text()")
for h1 in h1_tags:
    print(h1)

# Example: Extract all links (<a> tags)
links = tree.xpath("//a/@href")
for link in links:
    print(link)
```
