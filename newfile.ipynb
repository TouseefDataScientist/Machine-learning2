Droping Columns

```python
columns = ['Ticket', 'Cabin', 'Name']
```

```python
df = df.drop(columns, axis=1)
```

```python
df
```

Looking for null values

```python
df.isnull().sum()
```

Replacing nan value in Age by median

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df
```

replacing nan value in Embarked colummn by mode

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df
```

```python
df.isnull().sum()
```

Replacing female by "1" and male by "0" values in sex colummn

```python
df['Gender'] = df['Sex'].map({'female': 0, 'male': 1}).astype(int)
```

```python
df
```

Cmparing Age and fare graphically

```python
def comtitan():
    x=df['Age']
    y=df['Fare']
    plt.scatter(x, y, color='red', label='Age')
    plt.scatter(x, y, color='green', label='Fare', alpha=0.5)
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.title('scatter plot of Fare and Age')
    plt.legend(['Age(red)', 'Fare(green)'])
    plt.show()
```
