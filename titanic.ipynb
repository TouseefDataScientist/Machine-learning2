# Function

```python
def compare_age_fare(df):
    
    conditions = [
        (df['Age'] <= 12),  # Child
        (df['Age'] > 12) & (df['Age'] <= 18),  # Teenager
        (df['Age'] > 18) & (df['Age'] <= 60),  # Adult
        (df['Age'] > 60)  # Old
    ]
    age_categories = ['Child', 'Teenager', 'Adult', 'Old']

    df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, float('inf')], labels=age_categories)

    avg_fare_by_group = df.groupby('AgeGroup')['Fare'].mean()

    def compare_fare(row):
        group_avg_fare = avg_fare_by_group[row['AgeGroup']]
        if row['Fare'] > group_avg_fare:
            return 'Above Average'
        elif row['Fare'] < group_avg_fare:
            return 'Below Average'
        else:
            return 'Average'

    df['FareComparison'] = df.apply(compare_fare, axis=1)
    
    return df
```

# call Function

```python

df = pd.read_csv('titanic.csv')  

df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)

df = compare_age_fare(df)

print(df[['Age', 'Fare', 'AgeGroup', 'FareComparison']].head())
```
